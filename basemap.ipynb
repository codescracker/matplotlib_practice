```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8,8))

baseMap = Basemap(projection='merc', llcrnrlon=70, llcrnrlat=15,
        urcrnrlon=140, urcrnrlat=55, lat_0=15, lon_0=95, resolution='l')

baseMap.drawcoastlines()
baseMap.drawcountries()
baseMap.drawrivers(color= 'b')
baseMap.etopo()

lat, lon = [29.4316, 22.54315], [106.9123, 114.0579]

xpt, ypt = baseMap(lon, lat)
baseMap.plot(xpt, ypt, 'r-', markersize=15, linewidth = 5)
baseMap.plot(xpt, ypt, 'w*', markersize=15, linewidth = 5)

plt.show()
```

```python
# %load ./matplotlib_sourcecode/0506\ -\ plotting\ coordinates\ with\ basemap.py
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill',
            llcrnrlat = 20,
            llcrnrlon = -130,
            urcrnrlat = 50,
            urcrnrlon = -60,
            resolution = 'l')

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='black')

# 29 north 95 west
# Houston, TX
lat, lon = 29.7604, -95.3698
xpt, ypt = m(lon, lat)
m.plot(xpt, ypt, 'r*', markersize=15)

# Boulder, CO
lat, lon = 40.125, -104.237
xpt, ypt = m(lon, lat)
m.plot(xpt, ypt, 'bo')

land_check = m.is_land(xpt, ypt)

print(land_check)


plt.title('Basemap Example with Title')
plt.show()
```
